# Ordinal Encoding without using Sklearn OrdinalEncoder()


```python
import kagglehub
import pandas as pd
import os

# Download latest version
path = kagglehub.dataset_download("gcreatives/impact-of-covid-19-on-working-professionals")

print("Path to dataset files:", path)

# Get the CSV file path (assuming there's only one CSV file in the directory)
csv_file = [f for f in os.listdir(path) if f.endswith('.csv')][0]
csv_file_path = os.path.join(path, csv_file)

# Now read the CSV file
data = pd.read_csv(csv_file_path)
```

```python
data
```

```python
obj_cols = [x for x in data if data[x].dtype == "object"]
```

```python
obj_cols
```

```python
def ordinal_encoder(lst,columns):
  temp_lst = sorted(list(set(lst[columns])))
  encoding_dict = {val: i for i, val in enumerate(temp_lst)}
  # print(encoding_dict)
  #lst[columns + '_encoded'] = lst[columns].map({'Education': 55, 'Healthcare': 22, 'IT': 21, 'Retail': 3})
  lst[columns + '_encoded'] = lst[columns].map(encoding_dict)
  return lst
```

```python
encoded = ordinal_encoder(data,"Sector")
encoded
```

# **OneHotEncoder without using Sklearn**

```python
def onehot_encoder(lst,columns):
  temp_lst = sorted(list(set(lst[columns])))

  for value in temp_lst:
        lst[f'{columns}_{value}'] = (lst[columns] == value).astype(int)
  return lst
```

```python
encoded = onehot_encoder(data,"Sector")
encoded
```
